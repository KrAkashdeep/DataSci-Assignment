```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import warnings
from sklearn.preprocessing import StandardScaler,MinMaxScaler,RobustScaler
warnings.filterwarnings("ignore")
```

```python
hrdata=pd.read_csv('https://raw.githubusercontent.com/tkseneee/Dataset/dd7313e0c6487acd9ed7cd32b786782c7a1d3885/HR_data.csv')
hrdata.head(2)
```

```python
#1
hrdata.isnull().sum()
```

```python
#2
num_data=hrdata.select_dtypes(include=np.number)
# num_data.isnull().sum()
print(num_data.head(2))
```

```python
#3
num_data.max()-num_data.min()
```

```python
#4
capital_gain_col= num_data['Capital_Gain']
std_col=capital_gain_col.std()
mean_col=capital_gain_col.mean()
scaled=(capital_gain_col-mean_col)/std_col
print(scaled.head())


```

```python
#5
min_val=scaled.min()
max_val=scaled.max()
mean_val=scaled.mean()
std_val=scaled.std()

print("minimum value:-",min_val,"maximum value:-",max_val,"mean value:-",mean_val,"standard deviation:-",std_val)
```

```python
#6
scaler= StandardScaler()
scaled_data=scaler.fit_transform(num_data)
scaled_data=pd.DataFrame(scaled_data,columns=num_data.columns)
print(scaled_data.head())
```

```python
#7
scaled_data_mean=scaled_data.mean()
scaled_data_std=scaled_data.std()
print(scaled_data_mean)
print(scaled_data_std)
```

```python
#8
inverse_data=scaler.inverse_transform(scaled_data)
inverse_data=pd.DataFrame(inverse_data,columns=num_data.columns)
print(inverse_data.head())
```

```python
#9
min_max_scaler=MinMaxScaler()
scaled_data_mm=min_max_scaler.fit_transform(num_data)
scaled_data_mm=pd.DataFrame(scaled_data_mm,columns=num_data.columns)
(max_val_mm,min_val_mm)=(scaled_data_mm.max(),scaled_data_mm.min())
(max_val_mm,min_val_mm)
```

```python
#10
robust_scaler=RobustScaler()
scaled_data_rb=robust_scaler.fit_transform(num_data)
scaled_data_rb=pd.DataFrame(scaled_data_rb,columns=num_data.columns)
# (max_val_rb,min_val_rb)=(scaled_data_rb.max(),scaled_data_rb.min())
scaled_data_rb.head()
```
